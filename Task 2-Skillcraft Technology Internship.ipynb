#### <font color = lightgreen>**DATA CLEANING**</fonts>

##### <font color = lightblue>**Handle Missing values**</fonts>

```python
#Fill missing age with Median
df['Age'].fillna(df['Age'].median() , inplace=True)

#Drop cabin due to many missing values
df.drop(columns='Cabin', inplace = True)

#Fill Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# After cleaned the data
df.isnull().sum()
```

###### <font color = lightblue>**Encoding**</fonts>

```python
#Convert Categorical Variables
labelencoder=LabelEncoder()
df['Sex'] = labelencoder.fit_transform(df['Sex'])
```

```python
df['Embarked']=labelencoder.fit_transform(df['Embarked'])
```

```python
#Boxplot to find outliers
plt.figure(figsize=(10,5))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

##### <font color = lightgreen>**EXPLORATORY DATA ANALYSIS**</fonts>

##### <font color = lightblue>**UNIVARIATE ANALYSIS**</fonts>

```python
#Survival count
plt.figure(figsize=(10,5))
sns.countplot(x='Survived' ,data=df)
plt.title('Survival Count')
plt.show()
```

```python
#Passenger Class count
plt.figure(figsize=(10,5))
sns.countplot(x='Pclass' ,data=df)
plt.title('Passenger Class count')
plt.show()
```

```python
#Age distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Age'],kde=True,bins=30)
plt.title('Age Distribution')
```

```python
#Fare distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Fare'],kde=True,bins=30)
plt.title('Fare Distribution')
```

##### <font color = lightblue>**BIVARIATE ANALYSIS**</fonts>

```python
#Survival by sex
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.show()
```

```python
#Age Vs Survival
sns.histplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10,5))
num_cols=df.select_dtypes(include=['int64','float64'])
sns.heatmap(num_cols.corr(), annot=True, cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

##### <font color = lightblue>**MULTIVARIATE ANALYSIS**</fonts>

```python
#Pairplot
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass' ,'SibSp', 'Parch']], hue='Survived', palette='Set2')
plt.show()    
```
